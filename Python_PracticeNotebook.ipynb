```python
def flames(name1, name2):
    # Remove spaces and convert names to lower case
    name1 = name1.replace(" ", "").lower()
    name2 = name2.replace(" ", "").lower()

    # Create a list of the letters in each name
    name1_letters = [letter for letter in name1]
    name2_letters = [letter for letter in name2]

    # Remove common letters from the lists
    for letter in name1_letters:
        if letter in name2_letters:
            name1_letters.remove(letter)
            name2_letters.remove(letter)

    # Count the remaining letters and find the result using the FLAMES acronym
    total_letters = len(name1_letters) + len(name2_letters)
    if total_letters == 0:
        return "You are already friends."
    elif total_letters == 1:
        return "You are just good friends."
    elif total_letters == 2:
        return "You are lovers."
    elif total_letters == 3:
        return "You are engaged."
    elif total_letters == 4:
        return "You are getting married."
    elif total_letters == 5:
        return "You are siblings."
    elif total_letters == 6:
        return "You are enemies."

# Test the function
print(flames("Siddhartha", "Chetna "))
# print(flames("Emily", "Mike"))
# print(flames("Bob", "Carol"))
```

Binary Search Code:

```python
def bin_search(arr,k):
    n=len(arr);
    l=0;
    h=n-1;
   
    while l<=h:
        mid=(l+h)//2;
        if arr[mid]==k:
            return True;
        if arr[mid]<k:
            l=mid+1;
        if arr[mid]>k:
            h=mid-1;
    return False;

arr=[5,6,1,2,4,6,7,9,12];
k=9;
bin_search(arr,k)
```

Floor_Value of a number in sorted array:

```python
# import
def find_floor(arr,num):
    n=len(arr);
    l=0;
    h=n-1;
    # ans=sys.int_info.min
    ans= -100
    while l<=h:
        m=(l+h)//2;
        if arr[m]==num:
            return arr[m]
        if arr[m]>num:
            h=m-1;
        if arr[m]<num:
            ans=arr[m]
            l=m+1;
    return ans;        
    
arr=[1,2,4,6,7,9,12];
num=13;
find_floor(arr,num)
```

**Local Minima in an unsorted array**

```python
def local_minima(arr):
  n=len(arr)
  if arr[0]<arr[1]:
    return arr[0];
  if arr[n-1]<arr[n-2]:
    return arr[n-1];
  l=1;
  h=n-2;
  while(l<=h):
    m=(l+h)//2;
    if arr[m]<arr[m-1] and arr[m]<arr[m+1]:
      return arr[m];
    if arr[m]>arr[m+1]:
      l=m+1;
    else:
      h=m-1;

arr=[4,2,10,3,6,7,5,9,12];
local_minima(arr)


      

      
```
